### a.

```python
def con(a):
    if  a=='Completed':
        return 'Completed'
    elif a=='Defaulted' or a=='Chargedoff': 
        return 'Defaulted'
    elif a=='Defaulted' and a=='Chargedoff':
        return 'Defaulted'


s['LoanStatus']=s.apply(lambda x:con(x['LoanStatus']),axis=1)

```

```python
s['LoanStatus'].value_counts()
```

```python
s=s[s['LoanStatus'].notna()]
```

### b.

```python
categories = {1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other',11:'Other',13:'Other',14:'Other',15:'Other',16:'Other',18:'Other',19:'Other',20:'Other'}
def cat(a):
       if a in categories.keys():
            return(categories[a])
        
            

s['ListingCategory (numeric)']=s.apply(lambda x:cat(x['ListingCategory (numeric)']),axis=1)
```

```python
s['ListingCategory (numeric)'].value_counts()
```

```python
s=s[s['ListingCategory (numeric)'].notna()]
```

### c.
